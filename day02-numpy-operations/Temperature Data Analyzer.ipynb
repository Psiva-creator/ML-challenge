```python
import numpy as np

# Create random number generator with seed for reproducibility
rng=np.random.default_rng(seed=42)

# Generate 30 random temperatures between 15 and 45
temps=rng.integers(15,45,size=30)

# Print full temperature array
print("Monthly temperatures:",temps)

# Calculate and print basic statistics
print("Mean:",np.mean(temps))
print("Max temp:",np.max(temps))
print("Min temp:",np.min(temps))

# Find index of hottest and coldest day using argmax and argmin
print(f"Hottest day: Day {np.argmax(temps)+1} with {np.max(temps)}°C")
print(f"Coldest day: Day {np.argmin(temps)+1} with {np.min(temps)}°C")

# Boolean indexing to filter temperatures above mean
print("Above average temps:",temps[temps>np.mean(temps)])

# Count days where temperature was below 25 using boolean sum
print("Days below 25°C:",np.sum(temps<25))

# Reshape 30 days into 5 weeks of 6 days each
weekly_temps=temps.reshape(5,6)
print("Weekly temps:",weekly_temps)

# Calculate mean of each week using axis=1 (row-wise)
print("Weekly averages:",np.mean(weekly_temps,axis=1))

# Normalize temperatures between 0 and 1 using min-max formula
normalized=(temps-np.min(temps))/(np.max(temps)-np.min(temps))
print("Normalized temps:",normalized)

# Filter temperatures between 30 and 40 using two conditions with & operator
print("Temps between 30 and 40:",temps[(temps>=30) & (temps<=40)])
```
