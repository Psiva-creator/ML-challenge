```python
import pandas as pd

# ── Dataset ───────────────────────────────────────────────────────
data = {
    "Name":    ["Alice", "Ram", "Siva", "Raj", "Kavitha"],
    "Math":    [85, 62, 90, 78, 55],
    "Science": [76, 88, 72, 95, 60],
    "English": [90, 74, 85, 68, 80]
}

# Create DataFrame
df = pd.DataFrame(data)

# ── Calculations ──────────────────────────────────────────────────

# Total marks per student
df["Total"] = df[["Math", "Science", "English"]].sum(axis=1)

# Average marks per student
df["Average"] = df[["Math", "Science", "English"]].mean(axis=1)

# Assign grades based on average
df["Grade"] = df["Average"].apply(
    lambda x: "A" if x > 80 else ("B" if x > 70 else "C")
)

# ── Results ───────────────────────────────────────────────────────

# Top student
top_student = df["Name"][df["Average"].argmax()]

print("=" * 45)
print("        STUDENT MARKS REPORT")
print("=" * 45)
print(df.to_string(index=False))
print("=" * 45)
print(f"  Top Student : {top_student}")
print(f"  Class Avg   : {df['Average'].mean():.2f}")
print("=" * 45)


```
